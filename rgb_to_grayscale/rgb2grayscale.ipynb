```python
import os
from skimage.io import imsave,imread
from skimage.color import rgb2gray
```

```python
os.listdir()
```

```python
orig_dir = 'original'
files = [f for f in os.listdir(orig_dir) ]
```

```python
len(files)
```

```python
for image in files:
    print(image)
    original_image = imread(os.path.join(orig_dir,image))
    grayscale_image = rgb2gray(original_image)
    imsave(str('grayscale/'+image),grayscale_image)
```
